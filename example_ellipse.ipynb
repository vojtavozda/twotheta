# Ellipse

```python
import numpy as np
from numpy import sin, cos, tan, pi, sqrt
import elliptools as el
from matplotlib import pyplot as plt
import conetools
import importlib
from clrprint import printc
```

Let's define ellipse with following parameters

To the ellipse we can find a cone with angle `theta` which creates the ellipse
when intersecting `x-y` plane:

```python
ellipse = el.Ellipse(x0=0,y0=0,a=2,b=1.5,phi=0.1)
ellipse.print()
theta = 0.4 # Known from experiment
cone = ellipse.findCone(theta)
cone.print()
```

Plot the result

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

cone.plotMesh(ax,5)
ellipse.plot(ax)
ellipse.plotCone(ax)

ax.set_aspect('equal')
plt.show()
```
